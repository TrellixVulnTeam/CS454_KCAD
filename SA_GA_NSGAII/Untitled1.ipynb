```python
file1 = pd.read_pickle("bugzilla_eclipse_log(comments)_2016meancost.pkl")
file2 = pd.read_pickle("bugzilla_firefox_log(comments)_2016meancost.pkl")
file3 = pd.read_pickle("bugzilla_firefox_priority_2016meancost.pkl")
file4 = pd.read_pickle("bugzilla_firefox_comments+priority_2016meancost.pkl")
```

```python
file_list = [file1, file2, file3, file4]
bound_list = [50699.701903571, 262945.49580958224, 262945.49580958224, 262945.49580958224]

result = []

for i in range(len(file_list)):
    test_pkl = file_list[i]
    test_bound = bound_list[i]
    
    for j in range(3):
        test_NSGA2 = NSGA2.NRP_NSGA2(profit=test_pkl['Profit'], cost=test_pkl['Cost'], bound=test_bound, iteration=2500)
        temp = test_NSGA2.run_NSGA2()
        result.append(temp)
```
